# Linear Systems Visualizer

Visualize systems of linear equations in 2D.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_2d_system(equations, title="System of Linear Equations"):
    """
    Visualize a 2D system of linear equations.

    Args:
        equations: List of tuples (a, b, c) representing ax + by = c
        title: Plot title
    """
    fig, ax = plt.subplots(figsize=(10, 8))

    x = np.linspace(-10, 10, 400)
    colors = ['blue', 'red', 'green', 'purple', 'orange']

    for i, (a, b, c) in enumerate(equations):
        if abs(b) > 1e-10:
            # y = (c - ax) / b
            y = (c - a * x) / b
            ax.plot(x, y, color=colors[i % len(colors)],
                    label=f'{a}x + {b}y = {c}', linewidth=2)
        else:
            # Vertical line: x = c/a
            x_val = c / a
            ax.axvline(x=x_val, color=colors[i % len(colors)],
                       label=f'{a}x = {c}', linewidth=2)

    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(title)
    ax.legend()
    ax.grid(True, alpha=0.3)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.set_aspect('equal')

    return fig
```

## Solution Types

A system of two linear equations can have:
1. **Unique solution** - lines intersect at one point
2. **No solution** - lines are parallel (inconsistent)
3. **Infinite solutions** - lines are the same

```python
def visualize_solution_types():
    """Show all three types of solutions."""

    # Case 1: Unique solution (lines intersect)
    fig = plot_2d_system(
        [(2, 3, 8), (1, -2, -3)],
        "Unique Solution: Lines Intersect at (1, 2)"
    )
    # Mark intersection point
    plt.plot(1, 2, 'ko', markersize=10)
    plt.annotate('(1, 2)', (1, 2), textcoords="offset points",
                 xytext=(10, 10), fontsize=12)
    plt.show()

    # Case 2: No solution (parallel lines)
    fig = plot_2d_system(
        [(1, 1, 2), (1, 1, 3)],
        "No Solution: Parallel Lines"
    )
    plt.show()

    # Case 3: Infinite solutions (same line)
    fig = plot_2d_system(
        [(1, 1, 2), (2, 2, 4)],
        "Infinite Solutions: Same Line"
    )
    plt.show()
```

```python
visualize_solution_types()
```

## Gaussian Elimination Steps

Visualize how Gaussian elimination transforms the system:

Original system:
- 2x + 3y = 8
- x - 2y = -3

```python
def visualize_elimination_steps():
    """Show Gaussian elimination step by step."""
    # System:
    # 2x + 3y = 8
    #  x - 2y = -3

    steps = [
        ("Original System", [(2, 3, 8), (1, -2, -3)]),
        ("Swap rows (pivot)", [(1, -2, -3), (2, 3, 8)]),
        ("Eliminate: R2 - 2*R1", [(1, -2, -3), (0, 7, 14)]),  # 7y = 14
        ("Solve: y = 2", [(1, -2, -3), (0, 1, 2)]),
    ]

    for i, (title, eqs) in enumerate(steps):
        fig = plot_2d_system(eqs, f"Step {i+1}: {title}")
        plt.show()
```

```python
visualize_elimination_steps()
```
